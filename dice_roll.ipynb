```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.random.choice(a=[1,2,3,4,5,6],  p=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], size=1000)
plt.figure(figsize=(12,10))
plt.hist(x)
```

```
x = np.random.choice(a=[1,2,3,4,5,6], p=[0.15, 0.15, 0.15, 0.15, 0.15, 0.25], size=1000000)
plt.figure(figsize=(12,10))
plt.hist(x)
```

```
fig, ax_list = plt.subplots(ncols=2, nrows=1, figsize=(12,6))

x = np.random.choice(a=[1,2,3,4,5,6],  p=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], size=1000000)
ax_list[0].hist(x)

y = np.random.choice(a=[1,2,3,4,5,6], p=[0.15, 0.15, 0.15, 0.15, 0.15, 0.25], size=1000000)

ax_list[1].hist(y)
```
