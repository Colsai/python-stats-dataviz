# 11 - Discret Probability Distribution

Using np.random.choice() to simulate the rolling of a dice. It is a discret uniform distribution (the probability of each number is the same which is 1/6. 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulate the rolling of dice

rolls = np.random.choice(a=[1,2,3,4,5,6], size=10, replace=True, p=[1/6,1/6,1/6,1/6,1/6,1/6])
rolls
```

```python
rolls.mean()
```

```python
np.mean(rolls)
```

```python
# Roll the dice 300 times, each time, calculate the mean and plot the sample mean with the sample size

means = []
for i in range(1,300):
    rolls = np.random.choice(a=[1,2,3,4,5,6], size=i, replace=True, p=[1/6,1/6,1/6,1/6,1/6,1/6])
    means.append(rolls.mean())
fig, ax = plt.subplots(figsize=(12,8))
ax.set_ylim((1,6))
ax.plot(means)
```

```python
rolls = np.random.choice(a=[1,2,3,4,5,6], size=50, replace=True, p=[1/6,1/6,1/6,1/6,1/6,1/6])
rolls.mean()
```

```python
# This is to show how Central Limit Theorem works even when the underlying distribution is not normal. 
# Here the underlying distribution is a discret uniform distribution, but the distribution of the sample means is normal
# when the sample size is large.

sample_means = []
for i in range(0, 1000):
    rolls = np.random.choice(a=[1,2,3,4,5,6], size=500, replace=True, p=[1/6,1/6,1/6,1/6,1/6,1/6])
    sample_means.append(rolls.mean())
    
fig, ax = plt.subplots(figsize=(12,10))
ax.hist(sample_means)   
```
